```python
import pandas as pd

sampled_df = pd.read_csv('sampled_df.csv')

patients_table = sampled_df[['Name', 'Age', 'Gender', 'Blood Type']].drop_duplicates().reset_index(drop=True)
patients_table['Patient ID'] = ['p' + str(i + 1) for i in range(len(patients_table))]

medical_conditions_table = sampled_df[['Medical Condition']].drop_duplicates().reset_index(drop=True)
medical_conditions_table['Medical Condition ID'] = ['mc' + str(i + 1) for i in range(len(medical_conditions_table))]

admission_types_table = sampled_df[['Admission Type']].drop_duplicates().reset_index(drop=True)
admission_types_table['Admission Type ID'] = ['at' + str(i + 1) for i in range(len(admission_types_table))]

medications_table = sampled_df[['Medication']].drop_duplicates().reset_index(drop=True)
medications_table['Medication ID'] = ['m' + str(i + 1) for i in range(len(medications_table))]

test_results_table = sampled_df[['Test Results']].drop_duplicates().reset_index(drop=True)
test_results_table['Test Results ID'] = ['tr' + str(i + 1) for i in range(len(test_results_table))]

admissions_table = sampled_df.merge(patients_table, on=['Name', 'Age', 'Gender', 'Blood Type'], how='left')
admissions_table = admissions_table.merge(medical_conditions_table, on='Medical Condition', how='left')
admissions_table = admissions_table.merge(admission_types_table, on='Admission Type', how='left')
admissions_table = admissions_table.merge(medications_table, on='Medication', how='left')
admissions_table = admissions_table.merge(test_results_table, on='Test Results', how='left')

admissions_table = admissions_table[['Patient ID', 'Doctor', 'Hospital', 'Medical Condition ID', 
                                     'Date of Admission', 'Discharge Date', 'Billing Amount', 
                                     'Room Number', 'Admission Type ID', 'Medication ID', 
                                     'Test Results ID']].drop_duplicates().reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt

def plot_admission_types(results):
   
    admission_types = [item[0] for item in results]
    counts = [item[1] for item in results]
    
    plt.figure(figsize=(10, 6))
    plt.bar(admission_types, counts, color='skyblue')
    
    plt.title('Hospital Admissions by Type', fontsize=14, pad=20)
    plt.xlabel('Admission Type', fontsize=12)
    plt.ylabel('Number of Admissions', fontsize=12)

    for i, count in enumerate(counts):
        plt.text(i, count + 3, str(count), ha='center')
    
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    
    plt.show()

plot_admission_types(results)
```

```python
query = """
SELECT 
    medicalconditions.MedicalCondition AS Medical_Condition,
    DATEDIFF(admissions.DischargeDate, admissions.DateOfAdmission) AS LengthOfStay
FROM 
    admissions
JOIN 
    medicalconditions ON admissions.MedicalConditionID = medicalconditions.MedicalConditionID
WHERE 
    admissions.DischargeDate IS NOT NULL;  -- Ensures valid length of stay values
"""

# Execute the query
cursor.execute(query)

# Fetch and display results
results = cursor.fetchall()

print(results)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

def plot_medical_stay_heatmap(results):
    """
    Create a heatmap showing the distribution of length of stay across medical conditions.
    
    Parameters:
    results (list): List of tuples containing (medical_condition, length_of_stay)
    
    Returns:
    None: Displays the heatmap
    """
    df = pd.DataFrame(results, columns=['Medical_Condition', 'LengthOfStay'])

    stay_bins = [0, 3, 7, 14, 30, float('inf')]
    stay_labels = ['0-3 days', '4-7 days', '8-14 days', '15-30 days', '30+ days']
    df['StayCategory'] = pd.cut(df['LengthOfStay'], bins=stay_bins, labels=stay_labels)
    
    pivot_table = pd.crosstab(df['Medical_Condition'], df['StayCategory'])
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_table, 
                annot=True,  # Show numbers in cells
                fmt='d',     # Format as integers
                cmap='YlOrRd',  # Yellow to Red color scheme
                cbar_kws={'label': 'Number of Patients'})
    
    plt.title('Length of Stay Distribution by Medical Condition', pad=20)
    plt.xlabel('Length of Stay')
    plt.ylabel('Medical Condition')
    
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)

    plt.tight_layout()
    
    plt.show()
    
    print("\nSummary Statistics:")
    print(df.groupby('Medical_Condition')['LengthOfStay'].agg(['count', 'mean', 'median', 'min', 'max']).round(2))

plot_medical_stay_heatmap(results)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from decimal import Decimal

def plot_financial_trends(results):
    """
    Create a dual-axis line plot showing total and average values by month.
    
    Parameters:
    results (list): List of tuples containing (month, total_amount, average_amount)
    """
    df = pd.DataFrame(results, columns=['Month', 'Total', 'Average'])
    
    df['Total'] = df['Total'].astype(float)
    df['Average'] = df['Average'].astype(float)
    
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax2 = ax1.twinx()
    
    line1 = ax1.plot(df['Month'], df['Total'], color='blue', marker='o', 
                     linewidth=2, label='Total Amount')
    ax1.set_xlabel('Month')
    ax1.set_ylabel('Total Amount ($)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')
    
    ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M'))
    
    line2 = ax2.plot(df['Month'], df['Average'], color='red', marker='s',
                     linewidth=2, label='Average Amount')
    ax2.set_ylabel('Average Amount ($)', color='red')
    ax2.tick_params(axis='y', labelcolor='red')
    
    plt.xticks(df['Month'], range(1, 13))
    
    plt.title('Monthly Financial Trends', pad=20)
    
    lines = line1 + line2
    labels = [l.get_label() for l in lines]
    ax1.legend(lines, labels, loc='upper right')
    
    ax1.grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    plt.show()
    
    print("\nSummary Statistics:")
    summary = {
        'Total Amount': {
            'Mean': df['Total'].mean(),
            'Min': df['Total'].min(),
            'Max': df['Total'].max()
        },
        'Average Amount': {
            'Mean': df['Average'].mean(),
            'Min': df['Average'].min(),
            'Max': df['Average'].max()
        }
    }
    for metric, stats in summary.items():
        print(f"\n{metric}:")
        for stat_name, value in stats.items():
            if metric == 'Total Amount':
                print(f"{stat_name}: ${value:,.2f}")
            else:
                print(f"{stat_name}: ${value:,.2f}")

plot_financial_trends(results)
```

```python
query = """
SELECT 
    CASE
        WHEN Age BETWEEN 0 AND 18 THEN '0-18'
        WHEN Age BETWEEN 19 AND 35 THEN '19-35'
        WHEN Age BETWEEN 36 AND 50 THEN '36-50'
        WHEN Age BETWEEN 51 AND 65 THEN '51-65'
        WHEN Age > 65 THEN '65+'
    END AS Age_Group,
    COUNT(admissions.AdmissionID) AS Total_Admissions
FROM 
    admissions
JOIN 
    patients ON admissions.PatientID = patients.PatientID
GROUP BY 
    Age_Group
ORDER BY 
    Total_Admissions DESC;
"""


cursor.execute(query)

results = cursor.fetchall()

print(results)
```

```python
import matplotlib.pyplot as plt

def plot_age_distribution_donut(results):
    """
    Create a donut chart showing the distribution of age groups.
    
    Parameters:
    results (list): List of tuples containing (age_group, count)
    """
   
    age_groups = [item[0] for item in results]
    counts = [item[1] for item in results]

    total = sum(counts)
    percentages = [count/total * 100 for count in counts]

    labels = [f'{age}\n{count} ({percent:.1f}%)' 
             for age, count, percent in zip(age_groups, counts, percentages)]

    colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FF99CC']

    plt.figure(figsize=(10, 8))

    plt.pie(counts, 
           labels=labels,
           colors=colors,
           wedgeprops=dict(width=0.4, edgecolor='white'),
           startangle=90,
           pctdistance=0.85,
           labeldistance=1.1)

    plt.title('Patient Age Distribution', pad=20, size=14)

    plt.text(0, 0, f'Total\n{total}', ha='center', va='center', fontsize=12)

    plt.axis('equal')

    plt.figtext(1.1, 0.5, 'Summary:\n' + 
                '\n'.join([f'{age}: {count} patients' for age, count in results]),
                bbox=dict(facecolor='white', alpha=0.5))

    plt.show()

    print("\nAge Distribution Statistics:")
    print(f"Total Patients: {total}")
    for age_group, count in results:
        percentage = (count/total * 100)
        print(f"{age_group}: {count} patients ({percentage:.1f}%)")

plot_age_distribution_donut(results)
```
